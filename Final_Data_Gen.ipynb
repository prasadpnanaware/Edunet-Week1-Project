Create / Load the Dataset

```python
# Week2_Data_Gen.ipynb - run this cell
import numpy as np
import pandas as pd

np.random.seed(42)
n = 500

data = {
    'Voltage': np.round(np.random.uniform(3.0, 4.2, n), 3),
    'Current': np.round(np.random.uniform(0.2, 3.5, n), 3),
    'Temperature': np.round(np.random.uniform(20, 60, n), 2),
    'SOC': np.round(np.random.uniform(10, 100, n), 2)
}

def label_health(v, t, soc):
    # simple rule-based label for dummy data
    if v > 3.85 and t < 40 and soc > 50:
        return 'Good'
    elif v > 3.50 and t < 50:
        return 'Weak'
    else:
        return 'Replace'

data['Health'] = [label_health(v,t,s) for v,t,s in zip(data['Voltage'], data['Temperature'], data['SOC'])]
df = pd.DataFrame(data)
df.to_csv('battery_data_week2.csv', index=False)
df.head()
```

Evaluate and Visualize Results

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('battery_data_week2.csv')

# Simple statistics
print(df.describe())
print("\nBattery Health Distribution:")
print(df['Health'].value_counts())

# Visualization 1: Pairplot
sns.pairplot(df, hue='Health', corner=True)
plt.suptitle('EV Battery Data Visualization (Week 2)', y=1.02)
plt.show()

# Visualization 2: Health count
sns.countplot(data=df, x='Health')
plt.title('Distribution of Battery Health Classes')
plt.show()
```
